# Industrial Manufacturing World Model Demo

This notebook trains a small RSSM-inspired manufacturing world model on a synthetic process-window environment.

```python
from imwm.data import generate_world_model_dataset, split_dataset
from imwm.models.rssm import ManufacturingRSSM
from imwm.training import train_world_model
from imwm.evaluation import evaluate_world_model
from imwm.planning import CEMPlanner
from imwm.visualization import plot_process_window

arrays = generate_world_model_dataset(num_episodes=40, episode_length=35, history_length=8)
train_arrays, test_arrays = split_dataset(arrays)
model = ManufacturingRSSM(input_dim=arrays.input_dim, action_dim=arrays.action_dim, state_dim=arrays.state_dim)
report = train_world_model(model, train_arrays, epochs=10, verbose=True)
evaluate_world_model(model, test_arrays)
```

```python
history = test_arrays.sample_history(0)
planner = CEMPlanner(candidates=256, iterations=4, horizon=3)
action, planning_report = planner.plan(model, history)
planning_report
```

```python
plot_process_window(model, history, 'notebooks/process_window_demo.png', recommended_action=action)
```
